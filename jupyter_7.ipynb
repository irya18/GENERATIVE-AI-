```python
import nltk
import PyPDF2
import random
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('punkt')

def get_movie_type(text):
    text = text.lower()

    if "fight" in text or "battle" in text or "war" in text:
        return "Action"
    elif "love" in text or "romance" in text:
        return "Romance"
    elif "ghost" in text or "horror" in text or "fear" in text:
        return "Horror"
    elif "funny" in text or "comedy" in text:
        return "Comedy"
    elif "magic" in text or "fantasy" in text:
        return "Fantasy"
    else:
        return "General / Unknown"

def suggest_movies(movie_type):
    recommendations = {
        "Action": ["Mad Max: Fury Road", "John Wick", "Avengers: Endgame", "Gladiator"],
        "Romance": ["Titanic", "The Notebook", "La La Land", "Pride & Prejudice"],
        "Horror": ["The Conjuring", "It", "Insidious", "The Nun"],
        "Comedy": ["3 Idiots", "Hera Pheri", "The Mask", "Home Alone"],
        "Fantasy": ["Harry Potter", "The Lord of the Rings", "Doctor Strange", "Avatar"],
        "General / Unknown": ["Inception", "Interstellar", "The Dark Knight"]
    }

    return recommendations.get(movie_type, ["No suggestions available"])


pdf_path = input("Enter movie PDF file name (with .pdf): ")

try:
    pdf = open(pdf_path, "rb")
except:
    print("❌ File not found. Please check the name.")
    exit()

reader = PyPDF2.PdfReader(pdf)

text = ""

for page in reader.pages:
    if page.extract_text():
        text += page.extract_text()

movie_type = get_movie_type(text)

print("\n🎬 Movie Type Detected:", movie_type)
print("\n-----------------------------\n")

suggestions = suggest_movies(movie_type)

print("🎥 Suggested Movies:")
for movie in suggestions:
    print("-", movie)

print("\n-----------------------------\n")

sentences = nltk.sent_tokenize(text)

vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(sentences)

keywords = vectorizer.get_feature_names_out()

print("\nMovie type:", movie_type)
print("\n--------------------------------------\n")
```
