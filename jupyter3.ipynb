```python
df=pd.read_csv('/content/drive/MyDrive/Book4.csv')
df
```

```python
inputs = df.drop('Survived',axis='columns')
```

```python
target = df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
le_PassengerId = LabelEncoder()
le_Pclass = LabelEncoder()
le_Name= LabelEncoder()
le_Sex = LabelEncoder()
le_Age = LabelEncoder()
le_SibSp = LabelEncoder()
le_Parch = LabelEncoder()
le_Ticket = LabelEncoder()
le_Fare = LabelEncoder()
le_Cabin = LabelEncoder()
le_Embarked = LabelEncoder()
```

```python
inputs['PassengerId_n'] = le_PassengerId.fit_transform(inputs['PassengerId'])
inputs['Pclass_n'] = le_Pclass.fit_transform(inputs['Pclass'])
inputs['Name_n'] = le_Name.fit_transform(inputs['Name'])
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['SibSp_n'] = le_SibSp.fit_transform(inputs['SibSp'])
inputs['Parch_n'] = le_Parch.fit_transform(inputs['Parch'])
inputs['Ticket_n'] = le_Ticket.fit_transform(inputs['Ticket'])
inputs['Fare_n'] = le_Fare.fit_transform(inputs['Fare'])
inputs['Cabin_n'] = le_Cabin.fit_transform(inputs['Cabin'])
inputs['Embarked_n'] = le_Embarked.fit_transform(inputs['Embarked'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['PassengerId','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```
