```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/Book1.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
# Drop unwanted index column
new_df = new_df.drop(columns=['Unnamed: 0'])
```

```python
reg.fit(new_df, price)
```

```python
reg.predict(pd.DataFrame({'area': [3300]}))
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/Book2.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df['year\t'], df['Per capita income '], color='red', marker='+')
```

```python
new_df = df.drop('Per capita income ', axis='columns')
new_df
```

```python
Per_capita_income = df['Per capita income ']
Per_capita_income
```

```python
from sklearn import linear_model
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,Per_capita_income)
```

```python
reg.predict(pd.DataFrame({'year\t': [2020]}))
```
