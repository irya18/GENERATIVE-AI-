### Ensure Google Drive is mounted and the CSV file path is correct.

```python
from google.colab import drive
drive.mount('/content/drive')

# Make sure 'Book3.csv' exists at this path in your Google Drive
df=pd.read_csv('/content/drive/MyDrive/Book3.csv')
```

```python
inputs = df.drop('sgpa>9',axis='columns')
target = df['sgpa>9']
```

```python
from sklearn.preprocessing import LabelEncoder
le_attendence = LabelEncoder()
le_practical_submission = LabelEncoder()
le_internal_marks= LabelEncoder()
```

```python
inputs['attendence_n'] = le_attendence.fit_transform(inputs['attendence'])
inputs['practical_submission_n'] = le_practical_submission.fit_transform(inputs['practical submission'])
inputs['internal_marks_n'] = le_internal_marks.fit_transform(inputs['internal marks'])
```

```python
inputs_n = inputs[['attendence_n', 'practical_submission_n', 'internal_marks_n']]
```

### Defining and Fitting the Decision Tree Model

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
print('Model re-fitted successfully with 3 features!')
```

```python
print(f'Prediction for [[2,1,0]]: {model.predict([[2,1,0]])}')
```

```python
model.score(inputs_n,target)
```
